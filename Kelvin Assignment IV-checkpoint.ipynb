```python
#Services leading to Customer Churn
#We use correlation of churn with all the factors
# Then we Generate a Bar chart displaying correlation 
#Number 1, converting Churn to a quantifiable variable

TelcoChurn['Churn'].replace(to_replace='Yes', value=1, inplace=True)

TelcoChurn['Churn'].replace(to_replace='No',  value=0, inplace=True)
TelcoChurn.head()
```

```python
Telcodummies = pd.get_dummies(TelcoChurn)
```

```python
#Correlation Plot
plt.figure(figsize=(18,9))
Telcodummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```
