```python
#Importing the necessary Tools
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly
import plotly.express as px
import plotly.graph_objs as go
```

```python
#Importing the Data
IBMData=pd.read_csv(r'C:\Users\ADMIN\Desktop\ADS Folder\ADS Assignments\WA_Fn-UseC_-HR-Employee-Attrition.csv')
IBMData.head()
```

```python
#Analysis of distance from home by job role and attrition.
Analysis1= IBMData.groupby(["JobRole","Attrition"])["DistanceFromHome"].mean().unstack().fillna(0)
Analysis1
```

```python
#Generating Bar plot for the above
Analysis1.plot(kind='bar')
```

```python
#Comparison of Average Monthly income vs Education and Attrition
Analysis2=IBMData.groupby(["Education", "Attrition"])["MonthlyIncome"].mean().unstack().fillna(0)
Analysis2
```

```python
#Generating a Bar Plot of the above analysis
Analysis2.plot(kind='bar')
```
